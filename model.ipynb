```python
x = df.drop(["species"],axis=1)
y=df[["species"]]
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y, test_size=.10)
```

```python
dt=DecisionTreeClassifier()
```

```python
model=dt.fit(xtrain,ytrain)
model.score(xtrain,ytrain)
```
